```python
import pandas as pd

data = {
    'age': [25, 45, 35, 50, 23, 40, 60, 48],
    'gender': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'],
    'blood_pressure': [120, 140, 130, 150, 110, 135, 160, 145],
    'cholesterol': [200, 240, 180, 260, 170, 220, 280, 250],
    'diabetes': [0, 1, 0, 1, 0, 1, 1, 1]
}

df = pd.DataFrame(data)

df.to_csv('../data/healthcare.csv', index=False)

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('../data/healthcare.csv')

# Basic info
print(df.info())
print(df.describe())

# Check missing values
print(df.isnull().sum())
```

```python
# Age distribution
plt.hist(df['age'])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Blood pressure vs Cholesterol
plt.scatter(df['blood_pressure'], df['cholesterol'])
plt.xlabel("Blood Pressure")
plt.ylabel("Cholesterol")
plt.title("BP vs Cholesterol")
plt.show()

# Diabetes count
sns.countplot(x='diabetes', data=df)
plt.title("Diabetes Count")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Convert gender to numeric
df['gender'] = df['gender'].map({'M': 0, 'F': 1})

# Features and target
X = df[['age', 'gender', 'blood_pressure', 'cholesterol']]
y = df['diabetes']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = LogisticRegression()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('../data/healthcare.csv')

# Basic info
df.info()
df.describe()

# Plot 1: Age Distribution
plt.hist(df['age'])
plt.title("Age Distribution")
plt.show()

# Plot 2: BP vs Cholesterol
plt.scatter(df['blood_pressure'], df['cholesterol'])
plt.xlabel("Blood Pressure")
plt.ylabel("Cholesterol")
plt.show()

# Plot 3: Diabetes Count
sns.countplot(x='diabetes', data=df)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Convert gender
df['gender'] = df['gender'].map({'M': 0, 'F': 1})

X = df[['age', 'gender', 'blood_pressure', 'cholesterol']]
y = df['diabetes']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```
